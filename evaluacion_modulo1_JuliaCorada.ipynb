```python
class TiendaOnline:
    def __init__(self): #esto son los atributos
        self.inventario = [] #lista vacia donde guardar los productos
        self.ventas_totales = 0.0 #el total de las ventas acumuladas en float
#esto de aqui los métodos
    def agregar_producto(self, nombre, precio, cantidad):
        for producto in self.inventario: #busco si existe el producto
            if producto['nombre'] == nombre:
                producto['cantidad'] += cantidad #le sumo la cantidad dada a lo que ya tengo
                return
        #si el producto no está, lo añado con esto
        nuevo_producto = {'nombre': nombre,'precio': precio,'cantidad': cantidad}
        self.inventario.append(nuevo_producto)
```
