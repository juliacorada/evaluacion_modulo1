```python
class TiendaOnline:
    def __init__(self, nombre_tienda): 
        self.nombre_tienda = nombre_tienda
        self.inventario = []
        self.ventas_totales = 0.0 
        self.clientes = {}

    def agregar_producto(self, nombre, precio, cantidad):
        for producto in self.inventario: #para buscar si existe el producto
            if producto['nombre'] == nombre:
                producto['cantidad'] += cantidad 
                return
        
        nuevo_producto = {'nombre': nombre,'precio': precio,'cantidad': cantidad} #añado producto con esto, sin esto no se registra
        self.inventario.append(nuevo_producto)

    def ver_inventario(self): #muestra el inventario con los detalles
        print(f'Inventario de {self.nombre_tienda}:\n')
        for producto in self.inventario:
            print(f"Nombre : {producto['nombre']} , Precio : ${producto['precio']} , Cantidad : {producto['cantidad']}\n")
    
    def buscar_producto(self,nombre):
        for producto in self.inventario: #busco si existe el producto
            if producto['nombre'] == nombre:
                print(f"Nombre : {producto['nombre']}, Precio : ${producto['precio']}, Cantidad: {producto['cantidad']}")
                return
            
        print("Producto no disponible") #si no esta el producto, lo pongo fuera del if porque si no no imprime nada
    
    def actualizar_stock(self, nombre, cantidad):
        for producto in self.inventario:
            if producto['nombre'] == nombre:
                 nueva_cantidad = producto['cantidad'] + cantidad
                 
                 if nueva_cantidad < 0:  #por si tengo que quitar algo del stock porque se ha deteriorado o vendido por ejemplo
                    print("Error: no hay suficiente stock para realizar la operación")
                    return
                 
                 producto['cantidad'] += cantidad
                 
                 print(f"Stock actualizado: {producto['nombre']}, nueva cantidad: {producto['cantidad']}")
                 return
            
        print("producto no encontrado en el inventario")

    def eliminar_producto(self, nombre):
        for producto in self.inventario:
            if producto['nombre'] == nombre:
                self.inventario.remove(producto)
                print(f"Producto eliminado: {nombre}")
                return

        print("El producto no está en el inventario")        

    def calcular_valor_inventario(self):
        valor_total= 0 
        for producto in self.inventario:
            valor_total += producto['precio'] * producto['cantidad']

        print(f"el valor del inventario es {valor_total}")
        return 
    
    #bonus a partir de aqui
    def agregar_cliente(self, nombre, email):
        if nombre in self.clientes:
            print("El Cliente ya está registrado")
        else:
            self.clientes[nombre] = {"email": email, "compras": []}
            print(f"Cliente {nombre} añadido correctamente")

    def ver_clientes(self):
        if not self.clientes:
            print("no hay clientes registrados")
            return
        
        for nombre, datos in self.clientes.items():
            print(f"Nombre: {nombre}, Email: {datos['email']}")
```
